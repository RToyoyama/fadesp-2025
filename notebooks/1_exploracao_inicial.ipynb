```python
# Célula 1: Importando as bibliotecas necessárias
import pandas as pd
import numpy as np

# Configuração para exibir todas as colunas de um DataFrame
pd.set_option('display.max_columns', None)

print("Bibliotecas importadas com sucesso!")
```

```python
# Célula 2: Definindo os caminhos para os arquivos de dados
# Usando caminhos relativos, pois funciona em qualquer computador
# desde que a estrutura de pastas seja mantida.
path_inep = '../data/raw/MICRODADOS_ED_SUP_IES_2023.CSV'
path_cnpq = '../data/raw/Relatorio_de_dados_abertos_CNPq.csv'

print(f"Caminho para o arquivo do INEP: {path_inep}")
print(f"Caminho para o arquivo do CNPq: {path_cnpq}")
```

```python
# Célula 3: Carregando o dataset do Censo da Educação Superior (INEP)
# Dados do governo brasileiro frequentemente usam codificação 'latin-1' e ';' como separador.
try:
    df_inep = pd.read_csv(path_inep, sep=';', encoding='latin-1')
    print("Dataset do INEP carregado com sucesso!")
except FileNotFoundError:
    print(f"Erro: O arquivo não foi encontrado no caminho '{path_inep}'. Verifique o caminho e o nome do arquivo.")

# Exibindo as 5 primeiras linhas para uma primeira impressão
df_inep.head()
```

```python
# Célula 4: Carregando o dataset de Bolsas do CNPq

# ANÁLISE DO PROBLEMA: A saída da Célula 7 mostrou que os nomes das colunas são 'Unnamed: 0', 'Unnamed: 1', etc.
# Isso indica que há linhas extras no topo do arquivo CSV (títulos, linhas em branco) que confundiram o pandas.
# A solução é usar o parâmetro 'skiprows' para pular essas linhas.
# Pelas características do arquivo, um bom chute é que precisará pular as 5 primeiras linhas.

try:
    # Mantemos a correção do DtypeWarning usando os índices das colunas.
    dtype_cnpq_indices = {
        0: 'str',
        7: 'str',
        23: 'str',
        24: 'str'
    }

    # A CORREÇÃO PRINCIPAL: Adiciona skiprows=5 para pular as 5 primeiras linhas do arquivo.
    # O pandas então usará a linha seguinte (a 6ª linha) como o cabeçalho (nomes das colunas).
    df_cnpq = pd.read_csv(
        path_cnpq,
        sep=';',
        encoding='latin-1',
        skiprows=5, # <--- A CORREÇÃO PARA O KEYERROR ESTÁ AQUI
        dtype=dtype_cnpq_indices # <--- A CORREÇÃO PARA O DTYPEWARNING ESTÁ AQUI
    )

    print("Dataset do CNPq carregado com sucesso, pulando as linhas de cabeçalho extras!")

    # Vamos verificar as colunas para confirmar que agora estão corretas
    print("\nColunas do DataFrame CNPq (agora devem estar corretas):")
    print(df_cnpq.columns)


except FileNotFoundError:
    print(f"Erro: O arquivo não foi encontrado no caminho '{path_cnpq}'. Verifique o caminho e o nome do arquivo.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo do CNPq: {e}")

# Exibindo as 5 primeiras linhas para confirmar
df_cnpq.head()
```

```python
# Célula 5: Explorando o DataFrame do INEP
print("--- Informações sobre o DataFrame do INEP ---")
# .info() fornece um resumo conciso do DataFrame, incluindo tipos de dados e valores não nulos.
df_inep.info()

print("\n--- Dimensões do DataFrame do INEP ---")
# .shape retorna uma tupla com (número de linhas, número de colunas)
print(f"O DataFrame do INEP possui {df_inep.shape[0]} linhas e {df_inep.shape[1]} colunas.")

print("\n--- Estatísticas Descritivas de Colunas Numéricas (INEP) ---")
# .describe() gera estatísticas descritivas para as colunas numéricas.
df_inep.describe()
```

```python
# Célula 6: Explorando o DataFrame do CNPq
print("--- Informações sobre o DataFrame do CNPq ---")
# Resumo do DataFrame do CNPq
df_cnpq.info()

print("\n--- Dimensões do DataFrame do CNPq ---")
# Dimensões do DataFrame do CNPq
print(f"O DataFrame do CNPq possui {df_cnpq.shape[0]} linhas e {df_cnpq.shape[1]} colunas.")

print("\n--- Estatísticas Descritivas de Colunas Numéricas (CNPq) ---")
# Estatísticas do CNPq (pode não haver muitas colunas numéricas relevantes aqui)
df_cnpq.describe()
```
